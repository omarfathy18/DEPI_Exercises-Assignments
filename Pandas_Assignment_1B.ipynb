#### #7. Filter the data set. Create three sub DataFrames from the `position` column for `G`, `F`, and `C`.

```python
pos_G = nba_df[nba_df['pos'] == 'G']
pos_F = nba_df[nba_df['pos'] == 'F']
pos_C = nba_df[nba_df['pos'] == 'C']
```

#### #8.  Run `describe()` on these new DataFrames. Compare the mean field goals (the `fg` column) between positions.

```python
describe_F = pos_F.describe()
describe_F
```

```python
describe_C = pos_C.describe()
describe_C
```

```python
mean_G = pos_G['fg'].describe()['mean']
mean_F = pos_F['fg'].describe()['mean']
mean_C = pos_C['fg'].describe()['mean']
```
